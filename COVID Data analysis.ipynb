```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movement = pd.read_csv('Global_Mobility_Report.csv')
```

```python
movement.head()
```

```python
x = movement.date
y = movement.retail_and_recreation_percent_change_from_baseline
plt.plot(x,y)
plt.show()
```

```python
movement.info()
```

```python
mClean = movement.drop('sub_region_1', axis = 1)
mClean = mClean.drop('sub_region_2', axis = 1)
mClean.head()
```

```python
mClean.info()
```

```python
mClean = mClean.dropna()
mClean.info()
```

```python
x = mClean.date
y = mClean.retail_and_recreation_percent_change_from_baseline
plt.plot(x,y)
plt.show()
```

```python
corr_matrix = mClean.corr()
print(corr_matrix)
```

```python
from pandas.plotting import scatter_matrix
attributes = ['retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline', 
             'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline', 
              'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline']
scatter_matrix(mClean[attributes])
plt.show()
```
