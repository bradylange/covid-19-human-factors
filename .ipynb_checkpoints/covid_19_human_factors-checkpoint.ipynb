```python
# Import required libraries
import os
import pickle
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
# Load and Explore data
# =============================================================================
# Load COVID-19 global cases
covid_19_cases_data = pd.read_csv("data/covid_19_cases/covid_19_global_cases.csv")
# Load COVID-19 global mobility percent changes
mobility_data = pd.read_csv("data/covid_19_mobility/covid_19_global_mobility_report.csv")

# Explore COVID-19 global cases
print(covid_19_cases_data.shape, "\n\n")
print(covid_19_cases_data.columns, "\n\n")
print(covid_19_cases_data.head(), "\n\n")
print(covid_19_cases_data.tail(), "\n\n")
print(covid_19_cases_data.info(), "\n\n")
print(covid_19_cases_data.describe(), "\n\n")

# Explore COVID-19 global mobility percent changes
print(mobility_data.shape, "\n\n")
print(mobility_data.columns, "\n\n")
print(mobility_data.head(), "\n\n")
print(mobility_data.tail(), "\n\n")
print(mobility_data.info(), "\n\n")
print(mobility_data.describe(), "\n\n")
```

```python
# Preprocess Data
# =============================================================================
# Sum null values
print(pd.isnull(covid_19_cases_data).sum(), "\n\n")
print(pd.isnull(mobility_data).sum(), "\n\n")

# Drop unnecessary features ("popData2018" may be useful)
cases = covid_19_cases_data.drop(["day", "month", "year",
                                  "countryterritoryCode", "popData2018"], 
                                 axis = 1)
mobility = mobility_data.drop(["sub_region_2"], axis = 1)
mobility = mobility.dropna(axis = 0, subset = ["sub_region_1"])

imp_mean = SimpleImputer(strategy = "mean")
imp_mean.fit(mobility.iloc[:, 5:])
mobility.iloc[:, 5:] = imp_mean.transform(mobility.iloc[:, 5:])

mobility = mobility.groupby(
    ["country_region_code", "date"]
)[["retail_and_recreation_percent_change_from_baseline", "grocery_and_pharmacy_percent_change_from_baseline", 
   "parks_percent_change_from_baseline", "transit_stations_percent_change_from_baseline",
   "workplaces_percent_change_from_baseline", "residential_percent_change_from_baseline"]].mean().reset_index()

cases["dateRep"] = pd.to_datetime(cases["dateRep"])
mobility["date"] = pd.to_datetime(mobility["date"])

print(mobility.shape)
print(mobility.columns)

# Inner join COVID-19 cases and mobility
cases_mobility = cases.merge(
    mobility, how = "inner", 
    left_on = ["dateRep", "geoId"], 
    right_on = ["date", "country_region_code"]
).drop(["dateRep", "geoId", "country_region_code", "continentExp"], axis = 1)

cases_mobility["date"] = cases_mobility["date"].astype("object")

num_feat_names = ["deaths", "retail_and_recreation_percent_change_from_baseline",
                 "grocery_and_pharmacy_percent_change_from_baseline",
                 "parks_percent_change_from_baseline",
                 "transit_stations_percent_change_from_baseline",
                 "workplaces_percent_change_from_baseline",
                 "residential_percent_change_from_baseline"]
cat_feat_names = ["date", "countriesAndTerritories"]
target_names = ["cases"]

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, df):
        return df[self.attribute_names].values

num_pipeline = Pipeline(
    steps = [
        ("selector", DataFrameSelector(num_feat_names)),
        ("mean_imputer", SimpleImputer(strategy = "mean")),
        ("std_scaler", StandardScaler())
    ]
)

cat_pipeline = Pipeline(
    steps = [
        ("selector", DataFrameSelector(cat_feat_names))
#         ("one_hot_encoder", OneHotEncoder(sparse = False))
    ]
)

target_pipeline = Pipeline(
    steps = [
        ("selector", DataFrameSelector(target_names))
    ]
)

ml_pipeline = FeatureUnion(
    transformer_list = [
        ("numeric_pipeline", num_pipeline),
        ("categoric_pipeline", cat_pipeline),
        ("target_pipeline", target_pipeline)
    ]
)

c_m = ml_pipeline.fit_transform(cases_mobility)
c_m[:, 7] = c_m[:, 7].astype("str")
print(c_m.shape)
print(c_m)

# Sum null values
print(pd.isnull(c_m).sum() == 1)

# Total averages by country with given time spans
country_avg_c_m = pd.DataFrame(
    [
        (i, 
         c_m[c_m[:, 8] == i, 0].mean(), 
         c_m[c_m[:, 8] == i, 1].mean(),
         c_m[c_m[:, 8] == i, 2].mean(),
         c_m[c_m[:, 8] == i, 3].mean(),
         c_m[c_m[:, 8] == i, 4].mean(),
         c_m[c_m[:, 8] == i, 5].mean(),
         c_m[c_m[:, 8] == i, 6].mean(),
         c_m[c_m[:, 8] == i, 9].mean(),
        )
        for i in np.unique(c_m[:, 8])
    ],
    columns = [cat_feat_names[1]] + num_feat_names + target_names
)

# Note: Don't have China and surrounding countries mobility data
print(country_avg_c_m["countriesAndTerritories"].unique())

# Total global average per date
global_avg_date_c_m = pd.DataFrame(
    [
        (i, 
         c_m[c_m[:, 7] == i, 0].mean(), 
         c_m[c_m[:, 7] == i, 1].mean(),
         c_m[c_m[:, 7] == i, 2].mean(),
         c_m[c_m[:, 7] == i, 3].mean(),
         c_m[c_m[:, 7] == i, 4].mean(),
         c_m[c_m[:, 7] == i, 5].mean(),
         c_m[c_m[:, 7] == i, 6].mean(),
         c_m[c_m[:, 7] == i, 9].mean(),
        )
        for i in np.unique(c_m[:, 7])
    ],
    columns = [cat_feat_names[0]] + num_feat_names + target_names
)

print(global_avg_date_c_m.shape)

global_avg_date_c_m.to_csv("data/covid_19_cases_mobility/covid_19_avg_global_cases_mobility.csv", 
                           index = False)

train, test = train_test_split(global_avg_date_c_m, test_size = 0.1, shuffle = True)
print("Train Data Shape:\n{0}\n".format(train.shape))
print("Test Data Shape\n{0}\n".format(test.shape))

# Serialize the training and testing sets
modeling_path = os.path.join("data", "modeling")
if not os.path.isdir(modeling_path):
    os.makedirs(modeling_path)
modeling_pickle_path = os.path.join(modeling_path, "cases_mobility_train_test.pickle")
with open(modeling_pickle_path, "wb") as f:
    pickle.dump([train, test], f)

# Remove target variable
# x_train = np.delete(train, 8, axis = 1)
# y_train = train[:, 8]
# x_test = np.delete(test, 8, axis = 1)
# y_test = test[:, 8]
x_train = train.drop(["cases", "date"], axis = 1)
y_train = train[["cases"]]
x_test = test.drop(["cases", "date"], axis = 1)
y_test = test[["cases"]]
print("X Train Data Shape:\n{0}\n".format(x_train.shape))
print("Y Train Data Shape:\n{0}\n".format(y_train.shape))
print("X Test Data Shape:\n{0}\n".format(x_test.shape))
print("Y Test Data Shape:\n{0}\n".format(y_test.shape))
```

```python
# Analysis
# =============================================================================
# Linear Regression
linear_reg = LinearRegression(n_jobs = -1)
print("Linear Regression model has started training...")
# Compute Linear Regression 10-fold cross-validation score
linear_scores = cross_val_score(
    linear_reg, 
    x_train, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Linear Regression model has finished training...")

# Polynomial features
poly_features = PolynomialFeatures(degree = 2, include_bias = False)
x_train_poly = poly_features.fit_transform(x_train)

# Elastic Net Regression
elastic_net_reg = ElasticNet(alpha = 0.1, l1_ratio = 0.5, max_iter = 1000)
print("Elastic Net Regression model has started training...")
# Compute Elastic Net Regression 10-fold cross-validation score
elastic_net_scores = cross_val_score(
    elastic_net_reg, 
    x_train_poly, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Elastic Net Regression model has finished training...")

# Decision Tree Regression
tree_reg = DecisionTreeRegressor()
print("Decision Tree Regression model has started training...")
# Compute Decision Tree Regression 10-fold cross-validation score
tree_scores = cross_val_score(
    tree_reg, 
    x_train, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Decision Tree Regression model has finished training...")

# Random Forest Regression
forest_reg = RandomForestRegressor(n_jobs = -1, verbose = 2)
print("Random Forest Regression model has started training...")
# Compute Random Forest Regression 10-fold cross-validation score
forest_scores = cross_val_score(
    forest_reg, 
    x_train, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Random Forest Regression model has finished training...")

# Support Vector Regression
supp_vec_reg = SVR(verbose = 2)
print("Support Vector Regression model has started training...")
# Compute Random Forest Regression 10-fold cross-validation score
supp_vec_scores = cross_val_score(
    supp_vec_reg, 
    x_train, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Support Vector Regression model has finished training...")

# Gradient Boosting Regression
grad_boost_reg = GradientBoostingRegressor(verbose = 2)
print("Gradient Boosting Regression model has started training...")
# Compute Gradient Boosting Regression 10-fold cross-validation score
grad_boost_scores = cross_val_score(
    grad_boost_reg, 
    x_train, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Gradient Boosting Regression model has finished training...")

# Model selection
target_range = y_train.max() - y_train.min()
linear_avg_rmse = np.sqrt(-linear_scores).mean() / target_range
elastic_net_avg_rmse = np.sqrt(-elastic_net_scores).mean() / target_range
tree_avg_rmse = np.sqrt(-tree_scores).mean() / target_range
forest_avg_rmse = np.sqrt(-forest_scores).mean() / target_range
supp_vec_avg_rmse = np.sqrt(-supp_vec_scores).mean() / target_range
grad_boost_avg_rmse = np.sqrt(-grad_boost_scores).mean() / target_range

model_names = ["Linear Regression", "Elastic Net Regression", "Decision Tree Regression", 
               "Random Forest Regression", "Support Vector Regression", 
               "Gradient Boosting Regression"]
avg_rmses = np.array([linear_avg_rmse, elastic_net_avg_rmse, tree_avg_rmse, 
                      forest_avg_rmse, supp_vec_avg_rmse, grad_boost_avg_rmse])
print(["{:0.2%}".format(item) for item in np.squeeze(avg_rmses)])

plt.figure(figsize = (20, 10))
plt.bar(x = [item for item in range(1, avg_rmses.shape[0] + 1)], height = np.squeeze(avg_rmses))
plt.xticks([item for item in range(1, avg_rmses.shape[0] + 1)], model_names)
plt.ylim(ymax = 0.37)
plt.show()

# Fine tuning the model - robust for when search space is small
param_grid = [{"max_leaf_nodes": [10, 50, 200]}, {"max_depth": [1, 10, 100]}]
grid_search = GridSearchCV(grad_boost_reg, param_grid, cv = 10, scoring = "neg_mean_squared_error", n_jobs = -1)
grid_search.fit(x_train, y_train.squeeze())
print("Best Tuned Hyperparameters:\n{0}\n".format(grid_search.best_params_))

# Evaluation scores
print("Mean Test Score:\n{0}\n".format(np.sqrt(-grid_search.cv_results_["mean_test_score"])))

# Parameters
print("All Tuned Hyperparameters:\n{0}\n".format(grid_search.cv_results_["params"]))
print("Mean Test Score by Hyperparameter:")
cv_res = grid_search.cv_results_
for params, mean_score in zip(cv_res["params"], cv_res["mean_test_score"]):
    print(params, "==>", np.sqrt(-mean_score))
    
# RMSE
grad_boost_rmse_scores = np.sqrt(-grid_search.cv_results_["mean_test_score"])
print("RMSE:\n{0}\n".format(grad_boost_rmse_scores))
avg_cross_val_grad_boost_rmse = grad_boost_rmse_scores.mean()
print("Average Cross-Validation RMSE:\n{0}\n".format(avg_cross_val_grad_boost_rmse))

# Best Gradient Boosting Regression
best_grad_boost_reg = grid_search.best_estimator_
print("Gradient Boosting Regression model has started training...")
# Compute Random Forest Regression 10-fold cross-validation score
best_grad_boost_scores = cross_val_score(
    best_grad_boost_reg, 
    x_train, 
    y_train, 
    scoring = "neg_mean_squared_error", 
    cv = 10,
    verbose = 2,
    n_jobs = -1
)
print("Gradient Boosting Regression model has finished training...")

# RMSE
best_grad_boost_rmse_scores = np.sqrt(-best_grad_boost_scores)
print("\nRMSE:\n{0}\n".format(", ".join(["{:0.4f}".format(item) for item in best_grad_boost_rmse_scores])))
avg_cross_val_best_grad_boost_rmse = best_grad_boost_rmse_scores.mean()
print("Average Cross-Validation RMSE:\n{0}\n".format(avg_cross_val_best_grad_boost_rmse))

# RMSE normalized
best_grad_boost_avg_rmse = np.sqrt(-best_grad_boost_scores).mean() / target_range
print("RMSE Normalized:\n{:0.2%}\n".format(float(best_grad_boost_avg_rmse.values)))

# Predict new data
h_test = best_grad_boost_reg.predict(x_test)
test_mse = mean_squared_error(y_test, h_test)
test_rmse = np.sqrt(test_mse) / target_range
print("Test RMSE:\n{:0.2%}\n".format(float(test_rmse.values)))
```
